```python
import random
intents = {
            "greeting":["Hi! how are you?","Hello what's up"],
            "goodbye":["Great day ahead","Nice to see you"],
            "address":["Address of the hotel is near Lajpat Nagar New Delhi"],
            "Rooms":["AC Rooms and Non AC Rooms"],
            "Availability":["There are 20 AC Rooms and 30 Non AC Rooms are available"],
            "Price":["AC ROOM : 10000/day ,Non AC Room: 5000/day"]
}
```

```python
def detect_response(query):
  prediction = humingbird.Text.predict(
    text = query, # inferencing text
    labels=["greeting","goodbye",
            "address","Rooms",
            "Availability","Price"] # suply potential labels
  )
  max_score=0
  for val in prediction:
    if(val['score']>max_score):
      max_score=val['score']
      max_val = val['score']
      max_val_text = val['className']
  print(max_val_text)
  return random.choice(intents[max_val_text])

n = int(input("Enter the number of iterations: "))
for i in range(n):
    query = input("Enter your query (or type 'n' to exit): ")
    if query.lower() == 'n':
        break
    response = detect_response(query)
    print("Response:", response)
```

```python
intents = {
    "greeting": ["Hi there! How can I assist you today?", "Hello! What brings you here?"],
    "goodbye": ["Goodbye! Have a wonderful day!", "See you later! Take care!"],
    "menu": ["Our menu includes a variety of dishes: pasta, pizza, salads, and desserts."],
    "specials": ["Today's specials are Grilled Salmon, Mushroom Risotto, and Tiramisu for dessert."],
    "hours": ["We are open from 8 AM to 10 PM every day."],
    "reservation": ["You can make a reservation by calling us at 123-456-7890 or through our website."]
}
```

```python
def predict_label(query, labels):
    predictions = [{"className": label, "score": random.uniform(0, 1)} for label in labels]
    return predictions

def detect_response(query):
    labels = list(intents.keys())
    prediction = predict_label(query, labels)  # inferencing text

    max_score = 0
    max_val_text = ""

    for val in prediction:
        if val['score'] > max_score:
            max_score = val['score']
            max_val_text = val['className']

    print(f"Detected intent: {max_val_text} with score: {max_score}")
    return random.choice(intents[max_val_text])

n = int(input("Enter the number of iterations: "))
for i in range(n):
    query = input("Enter your query (or type 'n' to exit): ")
    if query.lower() == 'n':
        break
    response = detect_response(query)
    print("Response:", response)
```
