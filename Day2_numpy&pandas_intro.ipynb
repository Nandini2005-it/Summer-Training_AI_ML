```python
import pandas as pd
path = '/content/ign.csv'
data = pd.read_csv(path)
print(type(data))
print(data.shape)
```

```python
# access the scores of all games
score = data['score']
# average of the score
avg = score.mean()
mn = score.min()
mx = score.max()
var = score.var()
std = score.std()
print(f"variance: {var}")
print(f"average: {avg}")
print(f"maximum: {mx}")
print(f"minimum: {mn}")
print(f"standard deviation: {std}")
```

```python
score.describe()
```

```python
# how many years of data you have
year = data['release_year'].unique()
print("all years info", year)
```

```python
# how many games are there in each year
games_no = data['release_year'].value_counts()
print(games_no)
```

```python
# how many unique score phrases
phrase = data['score_phrase'].unique()
print(phrase)
```

```python
# how many pc games are there
platform = data ['platform'].unique()
print(platform)
```

```python
# how many pc games are there
ind = data['platform']=='PC'
newData = data[ind] #keep only those where ind is true
print("total no of pc games are ", newData.shape[0])
```

```python
ind = data['score_phrase']=='Amazing'
newData = data[ind]
print("total no of amazing games are", newData.shape[0])
```

```python
#how many pc games are amazing
newData = data[(data['platform']=='PC') & (data['score_phrase']=='Amazing')]
print("total no of pc games are ", newData.shape[0])
```
