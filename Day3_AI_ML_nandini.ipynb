```python
nleg = ndf[ndf['Legendary'] == True]
nnleg = ndf[ndf['Legendary'] == False]
```

```python
ndf.columns
```

```python
plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
plt.title("Legendary pokemons")
sns.barplot(data=nleg, x='Generation', y='Total')

plt.subplot(1, 2, 2)
plt.title("Non-Legendary pokemons")
sns.barplot(data=nnleg, x='Generation', y='Total')

plt.show()
```

```python
#compare the attack adn dfense and speed w.r.t their types
plt.figure(1,(22,6))
plt.subplot(1,3,1)
plt.title('Type 1 pokemons with attack')
plt.xticks(rotation = 90)
sns.barplot(data = nleg,x = 'Type 1',y = 'Attack')

plt.subplot(1,3,2)
plt.title('Type 1 pokemons with defense')
plt.xticks(rotation = 90)
sns.barplot(data = nleg,x = 'Type 1',y = 'Defense')

plt.subplot(1,3,3)
plt.title('Type 1 pokemons with Speed')
plt.xticks(rotation = 90)
sns.barplot(data = nleg,x = 'Type 1',y = 'Speed')



plt.figure(2,(22,6))
plt.subplot(1,3,1)
plt.title('Type 1 pokemons with attack')
plt.xticks(rotation = 90)
sns.barplot(data = nnleg,x = 'Type 1',y = 'Attack')

plt.subplot(1,3,2)
plt.title('Type 2 pokemons with defense')
plt.xticks(rotation = 90)
sns.barplot(data = nleg,x = 'Type 2',y = 'Defense')

plt.subplot(1,3,3)
plt.title('Type 2 pokemons with Speed')
plt.xticks(rotation = 90)
sns.barplot(data = nleg,x = 'Type 2',y = 'Speed')
```

```python
plt.figure(1,(22,6))
plt.title('Type 1 pokemons with attack')
sns.barplot(data = nleg,x = 'Type 1',y = 'Attack',label = 'legendary')
sns.barplot(data = nnleg,x = 'Type 1',y = 'Attack',label = 'nonlegendary')
plt.legend()
```

```python
nleg_g = nleg.groupby('Type 1')[['Attack', 'Defense', 'Speed']].mean().reset_index()

# Melt the data to plot
melted_legend = nleg_g.melt(id_vars='Type 1', var_name='Stat', value_name='Value')

# Plotting with seaborn
plt.figure(figsize=(10, 6))
sns.barplot(data=melted_legend, x='Type 1', y='Value', hue='Stat')
plt.title('Average Attack, Defense, and Speed of Legendary Pokmon by Type 1')
plt.xlabel('Type 1')
plt.ylabel('Average Value')
plt.legend(title='Statistic')
plt.show()
```
