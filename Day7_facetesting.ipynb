```python
import matplotlib.pyplot as plt
import matplotlib.image as mimg
from sklearn import svm,model_selection,metrics
import numpy as np
import pandas as pd
import joblib
```

```python
sam =  np.random.randint(1,8)

uid = int(input("Enter uid : "))
path = "/content/drive/MyDrive/Colab_Notebooks/orl_face/orl_face/u%d/%d.png"%(uid,sam)
im = mimg.imread(path)
# print(im.ndim)
plt.figure(1,(5,5))
plt.imshow(im,cmap='gray')
plt.axis('off')
plt.title("Query image")
```

```python
feat = im.reshape(1,-1)
model_path = "/content/drive/MyDrive/Colab_Notebooks/final_model.pkl"
model = joblib.load(model_path)
id = model.predict(feat)
print("User id predicted is :",int(id[0]))
path = "/content/drive/MyDrive/Colab_Notebooks/orl_face/orl_face/u%d/%d.png"%(id[0],1)

im_pred = mimg.imread(path)

plt.figure(1,(8,5))
plt.subplot(1,2,1)
plt.imshow(im,cmap='gray')
plt.axis('off')
plt.title("Query image")

plt.subplot(1,2,2)
plt.imshow(im_pred,cmap='gray')
plt.axis('off')
plt.title("predicted user")
```
