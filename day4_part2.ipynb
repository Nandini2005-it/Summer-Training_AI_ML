# CLUSTERING(KMEANS)


```python
cluster = 2
np.random.seed(1)
ndata = 0+(100-0)*np.random.random((500,2))
c1x,c1y = ndata[0,0],ndata[0,1]
c2x,c2y = ndata[1,0],ndata[1,1]
print(c1x)
```

```python
for _ in range(20):  # Use _ if the loop variable is not used
    cluster1x = []
    cluster1y = []

    cluster2x = []
    cluster2y = []

    for i in range(ndata.shape[0]):
        d1 = np.sqrt((c1x - ndata[i, 0]) ** 2 + (c1y - ndata[i, 1]) ** 2)
        d2 = np.sqrt((c2x - ndata[i, 0]) ** 2 + (c2y - ndata[i, 1]) ** 2)
        if d1 < d2:
            cluster1x.append(ndata[i, 0])
            cluster1y.append(ndata[i, 1])
        else:
            cluster2x.append(ndata[i, 0])
            cluster2y.append(ndata[i, 1])

    # Update centroids

    c1x = sum(cluster1x) / len(cluster1x)
    c1y = sum(cluster1y) / len(cluster1y)
    # if cluster2x:  # Avoid division by zero
    c2x = sum(cluster2x) / len(cluster2x)
    c2y = sum(cluster2y) / len(cluster2y)
    plt.plot(cluster1x,cluster1y,'ro')
    plt.plot(cluster2x,cluster2y,'co')
```

```python
c1x, c1y = ndata[0, 0], ndata[0, 1]
c2x, c2y = ndata[1, 0], ndata[1, 1]
c3x, c3y = ndata[2, 0], ndata[2, 1]
c4x, c4y = ndata[3, 0], ndata[3, 1]

for _ in range(20):  # Use _ if the loop variable is not used
    cl1x, cl1y = [], []
    cl2x, cl2y = [], []
    cl3x, cl3y = [], []
    cl4x, cl4y = [], []

    for j in range(ndata.shape[0]):
        d1 = np.sqrt((c1x - ndata[j, 0]) ** 2 + (c1y - ndata[j, 1]) ** 2)
        d2 = np.sqrt((c2x - ndata[j, 0]) ** 2 + (c2y - ndata[j, 1]) ** 2)
        d3 = np.sqrt((c3x - ndata[j, 0]) ** 2 + (c3y - ndata[j, 1]) ** 2)
        d4 = np.sqrt((c4x - ndata[j, 0]) ** 2 + (c4y - ndata[j, 1]) ** 2)

        if d1 <= d2 and d1 <= d3 and d1 <= d4:
            cl1x.append(ndata[j, 0])
            cl1y.append(ndata[j, 1])
        elif d2 <= d1 and d2 <= d3 and d2 <= d4:
            cl2x.append(ndata[j, 0])
            cl2y.append(ndata[j, 1])
        elif d3 <= d1 and d3 <= d2 and d3 <= d4:
            cl3x.append(ndata[j, 0])
            cl3y.append(ndata[j, 1])
        else:
            cl4x.append(ndata[j, 0])
            cl4y.append(ndata[j, 1])

    # Update centroids
    if cl1x:
        c1x, c1y = sum(cl1x) / len(cl1x), sum(cl1y) / len(cl1y)
    if cl2x:
        c2x, c2y = sum(cl2x) / len(cl2x), sum(cl2y) / len(cl2y)
    if cl3x:
        c3x, c3y = sum(cl3x) / len(cl3x), sum(cl3y) / len(cl3y)
    if cl4x:
        c4x, c4y = sum(cl4x) / len(cl4x), sum(cl4y) / len(cl4y)
plt.plot(cl1x, cl1y, 'ro')
plt.plot(cl2x, cl2y, 'bo')
plt.plot(cl3x, cl3y, 'go')
plt.plot(cl4x, cl4y, 'yo')
plt.show()
```
